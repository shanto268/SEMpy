```python
import cv2
import numpy as np
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb

def load_image(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    return img

def preprocess_image(img):
    img_blurred = cv2.GaussianBlur(img, (5, 5), 0)
    return img_blurred

def segment_image(img):
    threshold = threshold_otsu(img)
    img_binary = img > threshold
    img_closed = closing(img_binary, square(3))
    img_cleared = clear_border(img_closed)
    return img_cleared

def label_regions(img_segmented):
    img_labeled, num_regions = label(img_segmented, return_num=True)
    return img_labeled, num_regions

def visualize_segmented_image(img_orig, img_labeled, num_regions):
    image_label_overlay = label2rgb(img_labeled, image=img_orig, bg_label=0)
    cv2.imshow('Labeled Image', image_label_overlay)
    print(f'Number of regions detected: {num_regions}')
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def main():
    image_path = 'test.jpeg'
    img = load_image(image_path)
    img_preprocessed = preprocess_image(img)
    img_segmented = segment_image(img_preprocessed)
    img_labeled, num_regions = label_regions(img_segmented)
    visualize_segmented_image(img, img_labeled, num_regions)

if __name__ == '__main__':
    main()
```
