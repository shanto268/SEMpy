```python
import os
import cv2
import numpy as np
import pandas as pd

def extract_lines(image_path):
    image = cv2.imread(image_path)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define the red color range in HSV
    lower_red = np.array([0, 70, 50])
    upper_red = np.array([10, 255, 255])

    mask = cv2.inRange(hsv, lower_red, upper_red)
    lines = cv2.HoughLinesP(mask, 1, np.pi/180, 100, minLineLength=50, maxLineGap=10)

    return lines

def process_lines(lines):
    lengths = []
    angles = []

    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            length = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
            angle = np.arctan2(y2 - y1, x2 - x1) * 180 / np.pi

            lengths.append(length)
            angles.append(angle)

    return lengths, angles

def main(folder_path):
    data = []

    for file in os.listdir(folder_path):
        if file.endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
            image_path = os.path.join(folder_path, file)
            lines = extract_lines(image_path)
            lengths, angles = process_lines(lines)

            if len(lengths) > 0:
                for length, angle in zip(lengths, angles):
                    data.append([file, length, angle])

    df = pd.DataFrame(data, columns=['Image', 'Length (nm)', 'Angle (degrees)'])
    return df

folder_path = "/Users/shanto/Library/CloudStorage/GoogleDrive-shanto@usc.edu/Shared drives/LFL/Imaging/SEM/nbr38/dose_test_apr_18/"
df = main(folder_path)
print(df)
```

## Extract Red Stuff Only:

```python
def extract_red(image_path):
    image = cv2.imread(image_path)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define the red color range in HSV
    lower_red = np.array([0, 70, 50])
    upper_red = np.array([10, 255, 255])

    mask = cv2.inRange(hsv, lower_red, upper_red)
    red_only = cv2.bitwise_and(image, image, mask=mask)
    
    return red_only

def save_red_images(folder_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for file in os.listdir(folder_path):
        if file.endswith(('.png', '.jpg', '.jpeg', '.tif', '.tiff')):
            image_path = os.path.join(folder_path, file)
            red_only = extract_red(image_path)
            output_path = os.path.join(output_folder, file)
            cv2.imwrite(output_path, red_only)

output_folder = "/Users/shanto/LFL/sem_image_analysis/nbr_measurements"
save_red_images(folder_path, output_folder)
```

```python
import os
import sys
import cv2
import math
import numpy as np
import torch
from pytesseract import pytesseract, Output
from PIL import Image

# Set the tesseract executable path
pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'  # Update this path to your tesseract executable

# Import CRAFT functions
sys.path.append('./CRAFT-pytorch')
from craft import CRAFT
from imgproc import resize_aspect_ratio
from file_utils import saveResult

def angle_between_points(p1, p2):
    return math.degrees(math.atan2(p2[1] - p1[1], p2[0] - p1[0]))

def test_net(net, image, text_threshold, link_threshold, low_text, cuda, poly):
    # resize
    img_resized, target_ratio, size_heatmap = resize_aspect_ratio(image, square_size=1280, interpolation=cv2.INTER_LINEAR, mag_ratio=1.5)
    ratio_h = ratio_w = 1 / target_ratio

    # preprocessing
    x = normalize_mean_variance(img_resized)
    x = torch.from_numpy(x).permute(2, 0, 1)    # [h, w, c] to [c, h, w]
    x = x.unsqueeze(0)    # [c, h, w] to [b, c, h, w]

    if cuda:
        x = x.cuda()

    # forward pass
    with torch.no_grad():
        y, feature = net(x)

    # make score and link map
    score_text = y[0, :, :, 0].cpu().data.numpy()
    score_link = y[0, :, :, 1].cpu().data.numpy()

    # Post-processing
    boxes, polys = getDetBoxes(score_text, score_link, text_threshold, link_threshold, low_text, poly)

    # Coordinate adjustment
    boxes = adjustResultCoordinates(boxes, ratio_w, ratio_h)
    polys = adjustResultCoordinates(polys, ratio_w, ratio_h)
    for k in range(len(polys)):
        if polys[k] is None:
            polys[k] = boxes[k]

    return boxes, polys

input_folder = '/Users/shanto/LFL/sem_image_analysis/nbr_measurements_ocr/'
output_folder = os.path.join(input_folder, 'results')

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Load the pre-trained CRAFT model
craft_net = CRAFT()
craft_net.load_state_dict(torch.load('./CRAFT-pytorch/weights/craft_mlt_25k.pth', map_location=torch.device('cpu')))
craft_net.eval()

if torch.cuda.is_available():
    craft_net.cuda()

def process_image(image_path, output_path):
    # Load the image
    img = cv2.imread(image_path)

    # Apply CRAFT text detection
    bboxes, _ = test_net(craft_net, img, text_threshold=0.7, link_threshold=0.4, low_text=0.4, cuda=torch.cuda.is_available(), poly=True)

    for box in bboxes:
        # Calculate the angle of the line
        p1, p2, p3, p4 = box
        angle = angle_between_points(p1, p2)

        # Crop the region of interest (ROI) around the bounding box
        x1, y1 = np.min(box[:, 0]), np.min(box[:, 1])
        x2, y2 = np.max(box[:, 0]), np.max(box[:, 1])

        roi = img[y1:y2, x1:x2]

        # Extract the text from the ROI
        roi_gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        roi_thresh = cv2.adaptiveThreshold(roi_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
        custom_config = r'--oem 3 --psm 6'
        extracted_text = pytesseract.image_to_string(roi_thresh, config=custom_config)

        # Overlay the information on the original image
        cv2.polylines(img, [box.astype(np.int32).reshape((-1, 1, 2))], True, (0, 255, 0), 2)
        cv2.putText(img, f"{extracted_text}, {angle:.1f} deg", (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

    # Save the processed image
    cv2.imwrite(output_path, img)

# Process all images in the input folder
for file in os.listdir(input_folder):
    if file.endswith('.png') or file.endswith('.jpg'):
        input_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, file)
        process_image(input_path, output_path)
```

# Image Segmentation Meta

https://github.com/facebookresearch/segment-anything/blob/main/notebooks/predictor_example.ipynb

I have a folder of SEM images - "/Users/shanto/Library/CloudStorage/GoogleDrive-shanto@usc.edu/Shared drives/LFL/Imaging/SEM/nbr38/dose_test_apr_18/" . The images are grayscale with the bottom 80 pixels representing the scale and setting of the SEM which can be cropped out for analysis. The red colored objects on the picture represent the measurements of features of interest of the superconducting architecture shown in the image. In these images 370 pixels represent 400 nanometers. I want to segment the various features of each of these images using https://github.com/facebookresearch/segment-anything and also label the sizes of the segmented contours. Generate python code to do so

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure
from segment_anything.segment_anything import SegmentAnything

# Define the input and output paths
input_path = "/Users/shanto/Library/CloudStorage/GoogleDrive-shanto@usc.edu/Shared drives/LFL/Imaging/SEM/nbr38/dose_test_apr_18/"
output_path = "/Users/shanto/LFL/sem_image_analysis/nbr_measurements/meta"

# Define a function to preprocess the SEM images by cropping out the bottom 80 pixels
def preprocess(image):
    return image[:-80,:]

# Define a function to segment the features in the SEM images and label the sizes of the segmented contours
def segment_and_label(image):
    # Apply thresholding to convert the image to binary
    thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    # Find contours in the binary image
    contours = measure.find_contours(thresh, 0.5)

    # Create a blank image to draw the segmented contours on
    seg_image = np.zeros_like(image)

    # Iterate over the contours
    for contour in contours:
        # Convert the contour to integers
        contour = contour.astype(int)

        # Calculate the size of the contour in nanometers
        size = round(len(contour) * 400 / 370, 2)

        # Draw the contour on the segmented image
        cv2.drawContours(seg_image, [contour], 0, 255, 1)

        # Add the size label to the segmented image
        cv2.putText(seg_image, f"{size} nm", tuple(contour[0]), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255, 1)

    return seg_image

# Define the path for the checkpoint file
checkpoint_file = "weights/sam_vit_h_4b8939.pth"

# Create a SegmentAnything object with a checkpoint
sa = SegmentAnything(segment_and_label, checkpoint=checkpoint_file)

# Iterate over the SEM images in the input folder
for file_name in os.listdir(input_path):
    if file_name.endswith(".tif"):
        # Read the image
        image = cv2.imread(os.path.join(input_path, file_name), cv2.IMREAD_GRAYSCALE)

        # Preprocess the image
        image = preprocess(image)

        # Segment the features and label the sizes of the segmented contours
        seg_image = sa.segment(image)

        # Save the segmented image with size labels to the output folder
        cv2.imwrite(os.path.join(output_path, file_name), seg_image)
```

```python
def show_mask(mask, ax, random_color=False):
    if random_color:
        color = np.concatenate([np.random.random(3), np.array([0.6])], axis=0)
    else:
        color = np.array([30/255, 144/255, 255/255, 0.6])
    h, w = mask.shape[-2:]
    mask_image = mask.reshape(h, w, 1) * color.reshape(1, 1, -1)
    ax.imshow(mask_image)
    
def show_points(coords, labels, ax, marker_size=375):
    pos_points = coords[labels==1]
    neg_points = coords[labels==0]
    ax.scatter(pos_points[:, 0], pos_points[:, 1], color='green', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)
    ax.scatter(neg_points[:, 0], neg_points[:, 1], color='red', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)   
    
def show_box(box, ax):
    x0, y0 = box[0], box[1]
    w, h = box[2] - box[0], box[3] - box[1]
    ax.add_patch(plt.Rectangle((x0, y0), w, h, edgecolor='green', facecolor=(0,0,0,0), lw=2))    
```
